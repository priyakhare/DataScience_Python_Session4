```python
#                                     Session 4 Assignment 
    
    
#*************************************************************************************

# Program 1.1  : Program to Find Area of Triangle using herons formula 
#                area  =  (s(s-a)*(s-b)*(s-c))**0.5

#*************************************************************************************

# Parent Class  : Triangle having functions , function  set_length for setting the lengths from user
#               : Another for printing the object as string            

class Triangle:
    
    def set_length(self,lenA,lenB,lenC):
        self._side1 = int(lenA)
        self._side2 = int(lenB)
        self._side3 = int(lenC)

    
    def __str__(self):
        return "The lengths of the sides of Triangle are %d ,%d, %d " %(self._side1,self._side2,self._side3)

    
    
#Importing decimal for using rounding feature for Area Value
from  decimal import Decimal

# SubClass Triangle Area is subclass of Parent Triangle , which calculates the Area 
# Function calc_area for calculating area using herons formula and returning the value with precision 2  
class TriangleArea(Triangle):
    
    def __init__(self):
        super(TriangleArea,self).__init__()
    
    def calc_area(self):
        self._perimeter = (self._side1 + self._side2 + self._side3)/2
        sidePer1 = self._perimeter - self._side1
        sidePer2 = self._perimeter - self._side2
        sidePer3 = self._perimeter - self._side3
        area = round(Decimal(((sidePer1 * sidePer2 * sidePer3 )*self._perimeter)**0.5),2)
        return area
    
    
    def __str__(self):
        return super(TriangleArea, self).__str__() +" Subclass create %d" %TriangleArea.calc_area(self)

     
# While Loop for taking the Lengths from user and creating list    
numofsides = 0
l = []
while numofsides<=2:
    l.append(input("please Enter the Triangle side lengths "))
    numofsides+=1

    
# Instantiating Object of subclass TriangleArea    
shapeArea = TriangleArea()
shapeArea.set_length(l[0],l[1],l[2])

print("The Area of Triangle is : " + str(shapeArea.calc_area()) + " cm2")


#*************************************************************************************
```

```python
# *************************************************************************************

# 1.2  Program  : To Write a function filter_long_words() that takes a list of words 
#                 and an integer n and returns the list of words that are longer than n. 

# *************************************************************************************

# Function to take list of words and filter only words longer than the size given
def filter_Long_words(lst,n):
    l = []
    for word in lst:
        if (len(word) > n):
            l.append(word)
    return l

# setting the list that needs to be filtered on size 
lst = ['a,sd','sdsd','asdasdasd','s','sds','dsfdddd','sdfsdfsdfsdfsdfsdf']
size = 4
# Invoking function to filter the words based on n paramter
lstfiltered = filter_Long_words(lst,size)
print("The Filtered long words in list are " + str(lstfiltered))    
    

# *************************************************************************************
```

```python
# 2.1  Write a Python program using function concept that maps  list of words into a list of integers representing the lengths of the corresponding words​. 
#  Hint: ​If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4] 
#  Here 2,3 and 4 are the lengths of the words in the list. 
# *************************************************************************************

#    Program 2.1 : To Write a Python program using function concept that maps  
#                  list of words into a list of integers representing the lengths of the 
#                  corresponding words. 

# *************************************************************************************


def words_to_size(lst):
    l = []
    for word in lst:
        l.append(len(word))
    return l


lst = ['sd','dfdsfsdf','dsfds','ddsfsfsd','dsfsddsdsf','dsf','dfdsfsfdfsfsdff']

lstsize = list(words_to_size(lst))
print("The mapping of words to size of \n {} is : \n   {}".format(lst,lstsize))        


# *************************************************************************************

```

```python
# 2.2  Write a Python function which takes a character (i.e. a string of length 1) and returns True 
# if it is a vowel, False otherwise. 
# *************************************************************************************

#   Program 2.2  :  Program that checks if the character is vowel

# *************************************************************************************
# Function to check if the character passed to it is a vowel 
# Character is checked if part of Vowels list
def check_vowel(char):
    vowels = "aeiou"
    if char in vowels:
        return True
    else:
        return False
    
    
# Validation to check user enters only alphabet character
while (True) :    
    checkChar = input("pls enter an alphabet  " )
    if((len(checkChar) == 1) and (checkChar.isalpha())):
        break
        
# print the Output based on the result of the check_vowel function        
if(check_vowel(checkChar)):
    print("The Character {} is a Vowel".format(checkChar))
else:
    print("The Character {} is not  a Vowel".format(checkChar))
    

# *************************************************************************************
```
